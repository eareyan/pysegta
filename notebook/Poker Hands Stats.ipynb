```python
def compute_conf_interval(sample, confidence_level=0.95):
    degrees_freedom = len(sample) - 1
    sample_mean = np.mean(sample)
    sample_standard_error = scipy.stats.sem(sample)
    lb, ub = scipy.stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)

    return sample_mean, lb, ub, (ub - lb) / 2
```

```python
data = pd.read_csv('../poker_experiments_data/all_results_straight_flush.csv')
pprint.pprint([compute_conf_interval(data[str(i)]) for i in [0, 1, 2, 3]])
```

## Inspect Results

```python
def combine_results(expt_name):
    games = pd.read_csv(f"../poker_experiments_data/experiment_{expt_name}_games.csv")
    psp_results = pd.read_csv(f"../poker_experiments_data/experiment_{expt_name}_psp.csv")
    psp_results['total'] = psp_results['psp_iter_0'] + psp_results['psp_iter_1'] + psp_results['psp_iter_2'] + psp_results['psp_iter_3']
    psp_means = psp_results.groupby('game_id').mean()
    psp_means = psp_means.reset_index()
    all_results = psp_means.merge(games)
    return all_results
```

```python
all_results = pd.concat([combine_results(expt_name) for expt_name in ['Willie_Namm', 
                                                                      'Terri_Newman',
                                                                      'David_Empie']])
len(all_results)
```

```python
all_results.sort_values('max_variance')
```

```python
xlabel = f"\n  \n eps = 0.15, \n failure schedule = [0.025, 0.025, 0.025, 0.025], \n sample schedule = [154, 308, 615, 1229]. \n {len(all_results)} samples"
```

```python
plt.scatter(all_results['max_variance'], all_results['total'], marker="x")
plt.xlabel(f"Variance (max) {xlabel}")
plt.ylabel('Total active (p, s) pairs')
plt.title('Total active pairs as a function of variance (max)')
```

```python
m, b = np.polyfit(all_results['max_variance'], all_results['total'], 1)
plt.scatter(all_results['max_variance'], all_results['total'], marker="x")
plt.plot(all_results['max_variance'], m*all_results['max_variance'] + b)
plt.xlabel(f"Variance (max) {xlabel}")
plt.ylabel('Total active (p, s) pairs')
plt.title('Total active pairs as a function of variance (max)')
```
